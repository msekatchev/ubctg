```python
x_len =  50
y_len = 1000

# Generate a random array of 0s and 1s
steps = np.random.randint(2, size=(x_len, y_len))

# Convert 0s to -1 and 1s to +1
steps[steps == 0] = -1

plt.figure(dpi=300)

p_val_walk  = np.zeros(x_len)
p_val_steps = np.zeros(x_len)

for x in range(x_len):
    plt.subplot(2, 1, 1)
    plt.plot(np.cumsum(steps[x]), "-", linewidth=0.5, alpha=0.6)
    p_val_walk[x]  = adfuller(np.cumsum(steps[x]))[1]

    plt.subplot(2, 1, 2)
    plt.plot(steps[x], "-", linewidth=0.5, alpha=0.2)
    p_val_steps[x] = adfuller(steps[x])[1]
```

```python
plt.figure(dpi=300)


plt.subplot(1,2,1)
plt.hist(p_val_walk, bins=20)
plt.title("walks (cumulative sums)")
plt.xlabel("p-value")

plt.subplot(1,2,2)
plt.hist(p_val_steps, bins=20)
plt.title("individual steps")
plt.xlabel("p-value")


plt.show()
```
